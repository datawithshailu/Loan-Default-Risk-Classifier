```python
import pandas as pd

# Step 2: Load the CSV from full file path
file_path = r"C:\Users\PREETHIKA RAVIRAJ\Downloads\archive (1)\loan_default.csv"
df = pd.read_csv(file_path)

# Show top 5 rows
df.head()

```

```python
# Step 3: Basic Data Cleaning

# Check column names and data types
print("ðŸ”¹ Column Info:")
print(df.info())

# Check for missing values
print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())

# Check for duplicate rows
print("\nðŸ”¹ Duplicates:")
print(df.duplicated().sum())
```

```python
# Convert age to numeric (some values may be strings)
df['age'] = pd.to_numeric(df['age'], errors='coerce')

# Now check again how many are null
print("Missing in 'age':", df['age'].isnull().sum())
```

```python
# Let's inspect the first few values in 'age'
df['age'].unique()[:20]
```

```python
# Drop the age column entirely
df.drop(columns=['age'], inplace=True)
```

```python
df['loan_limit'] = df['loan_limit'].fillna('Unknown')
df['approv_in_adv'] = df['approv_in_adv'].fillna('Unknown')
df['loan_purpose'] = df['loan_purpose'].fillna('Unknown')
df['Neg_ammortization'] = df['Neg_ammortization'].fillna('No')
df['term'] = df['term'].fillna(df['term'].median())
df['submission_of_application'] = df['submission_of_application'].fillna('Unknown')
```

```python
num_cols = ['property_value', 'income', 'LTV']  # removed 'age'

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# These columns have lots of missing data, so we drop the rows where theyâ€™re missing
df.dropna(subset=['rate_of_interest', 'Interest_rate_spread', 'Upfront_charges', 'dtir1'], inplace=True)
```

```python
features = ['income', 'Credit_Score', 'loan_amount', 'loan_purpose']
target = 'Status'

# Make a copy of just these columns
model_data = df[features + [target]].copy()
```

```python
model_data = pd.get_dummies(model_data, columns=['loan_purpose'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X = model_data.drop('Status', axis=1)
y = model_data['Status']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```
